```python
# reads processed electron spectra
s_sum = []
for n,row in gammaProf_db.dataframe.iterrows():
    run_name,shot = row.name
    spectra_path = Path('/data/analysis/GEMINI/2021/App20110008-1/Results/Espec/Spectra') / run_name.replace('/','_')
    espec_file = spectra_path / f'Shot{shot:04}_spectra.pkl'
    if os.path.isfile(espec_file):
        espec_data = load_object(espec_file)
        print(espec_data.keys())
        E_MeV,y_mrad,spec1 = espec_data['E_MeV'],espec_data['y1_mrad'],espec_data['spec1']
        #print(len(espec_data['dWdE1_JperGeV']))
        s_sum.append(np.trapz(np.trapz(spec1,x=y_mrad,axis=0),x=E_MeV))
    else:
        s_sum.append(0.0)
```

```python
from PIL import Image
from scipy.ndimage import median_filter

def load_img(file_path):
    return np.array(Image.open(file_path)).astype(float)
bkg_img = 0
def prep_img_file(file_path):
    img = load_img(file_path)-bkg_img
    img = median_filter(img,9)
    
    return img
```

```python
# read electron divergence and spectrum (and energy axis) for hits
# computes interpolating spline/smoothing function (to remove effect of wires) across spec and div (1d) data.
# shifts divergence so that centre lies at 0
plt.style.use('/home/el1103292/GitHub/RadiationReaction2021Analysis/lib/thesis.mplstyle')
normalisation_factor_E=np.random.uniform(800.0, 1500.0)

def compute_percentile_from_dist(dist, axis, target_percentile):
    
    div=10.0
    N=int(len(dist)/div)
    percentile, energy=np.zeros(N), np.zeros(N)
    dist=dist/np.trapz(dist, axis)
    for i in range(0, N):
        percentile[i]=np.trapz(dist[0:len(axis)-1-int(div)*i], axis[0:len(axis)-1-int(div)*i])
        energy[i]=axis[len(axis)-1-int(div)*i]
        if percentile[i]<target_percentile-0.005:
            break
    percentile_cut=percentile[percentile!=0.0]
    energy_cut=energy[energy!=0.0]
    energy_at_percentile=np.interp(target_percentile, percentile_cut, energy_cut)
    return energy_at_percentile

def compute_peakiness_measure(dQ_dE, E_MeV, color, normalisation_factor_E):
    E_GeV=E_MeV/1000.0
    dQ_dE=dQ_dE*E_GeV/np.trapz(dQ_dE*E_GeV, E_GeV)
    Q_tot=np.trapz(dQ_dE, E_GeV)
    mean_energy_tmp=np.trapz(dQ_dE*E_GeV, E_GeV)/Q_tot
    mean_energy_sqrd_tmp=np.trapz(dQ_dE*E_MeV*E_GeV, E_GeV)/Q_tot
    
    e_at_70th_percentile=compute_percentile_from_dist(dQ_dE, E_GeV, 0.7)
    e_at_50th_percentile=compute_percentile_from_dist(dQ_dE, E_GeV, 0.5)
    E_thresh=e_at_70th_percentile
    indx_near_mean=np.argmin(abs(E_GeV-E_thresh))

    mean_dQ_dE=np.mean(dQ_dE)
    max_over_med_above_thresh=max(dQ_dE[indx_near_mean:])#(max(dQ_dE[indx_near_mean:])-mean_dQ_dE)/mean_dQ_dE
    peakiness_measure=max_over_med_above_thresh
    return peakiness_measure, E_thresh, max(dQ_dE[indx_near_mean:])#peakiness_measure, mean_energy_tmp
    
    
np.random.seed(67)
rand_norm=np.random.uniform(0.1, 10.0)
e_img_list = []
E_90th_percentile_hits=[]
Qgamma2_hits=[]
av_gamma2_hits=[]
peakiness_hits_arr=[]
Q_energy_thresh_hits=[]
electron_total_KE_hits=[]
color=['b', 'r']

for ind in i_sel:#[2:3]
    row = gammaProf_db.dataframe.iloc[ind]
    run_name,shot = row.name

    spectra_path = Path('/data/analysis/GEMINI/2021/App20110008-1/Results/Espec/Spectra') / run_name.replace('/','_')
    espec_file = spectra_path / f'Shot{shot:04}_spectra.pkl'
    if os.path.isfile(espec_file):
        espec_data = load_object(espec_file)
        E_MeV,y_mrad,spec1, dWdE1_JperGeV = espec_data['E_MeV'],espec_data['y1_mrad'],espec_data['spec1'],espec_data['dWdE1_JperGeV']
        y0 = y_mrad[np.argmax(np.sum(spec1,axis=1))]
        y_mrad_disp = np.linspace(-4.5,4.5,100)
        spec_img = RectBivariateSpline(y_mrad-y0,E_MeV,spec1)(y_mrad_disp,E_MeV)
        e_img_list.append(spec_img)
        e_at_90th_percentile=compute_percentile_from_dist(dWdE1_JperGeV, E_MeV/1000.0, 0.9)
        E_90th_percentile_hits.append(e_at_90th_percentile)
        dQ_dE=np.trapz(spec1,x=y_mrad,axis=0)
        Q_tot=np.trapz(dQ_dE, E_MeV)
        Qgamma2_hits_tmp=np.trapz(dQ_dE*E_MeV/0.511*E_MeV/0.511, E_MeV)
        
        electron_total_KE_tmp=np.trapz(dQ_dE*(E_MeV/0.511-1.0), E_MeV/0.511)/Q_tot
        electron_total_KE_hits.append(electron_total_KE_tmp)
        
        Qgamma2_hits.append(Qgamma2_hits_tmp)
        av_gamma2_hits.append(Qgamma2_hits_tmp/Q_tot)
        peakiness_hits, energy_thresh_tmp, max_height_hits=compute_peakiness_measure(dQ_dE, E_MeV, color[1], normalisation_factor_E)
        Q_energy_thresh_hits.append(energy_thresh_tmp)
        peakiness_hits_arr.append(peakiness_hits)
    else:
        e_img_list.append(spec_img*np.nan)
        E_90th_percentile_hits.append(np.nan)
# and for nulls
e_null_list = []
E_90th_percentile_nulls=[]
Qgamma2_nulls=[]
av_gamma2_nulls=[]
Q_energy_thresh_nulls=[]
peakiness_nulls_arr=[]
electron_total_KE_nulls=[]
count=0
color_arr=['r', 'b', 'g', 'orange', 'cyan', 'magenta']
for lists in i_null:#[2:3]
    for ind in lists:#[0:1]
        row = gammaProf_db.dataframe.iloc[ind]
        run_name,shot = row.name

        spectra_path = Path('/data/analysis/GEMINI/2021/App20110008-1/Results/Espec/Spectra') / run_name.replace('/','_')
        espec_file = spectra_path / f'Shot{shot:04}_spectra.pkl'
        if os.path.isfile(espec_file):
            espec_data = load_object(espec_file)
            E_MeV,y_mrad,spec1,dWdE1_JperGeV = espec_data['E_MeV'],espec_data['y1_mrad'],espec_data['spec1'],espec_data['dWdE1_JperGeV']
            y0 = y_mrad[np.argmax(np.sum(spec1,axis=1))]
            y_mrad_disp = np.linspace(-4.5,4.5,100)
            spec_img = RectBivariateSpline(y_mrad-y0,E_MeV,spec1)(y_mrad_disp,E_MeV)
            e_null_list.append(spec_img)
            e_at_90th_percentile=compute_percentile_from_dist(dWdE1_JperGeV, E_MeV/1000.0, 0.9)
            E_90th_percentile_nulls.append(e_at_90th_percentile)
            dQ_dE=np.trapz(spec1,x=y_mrad,axis=0)
            Q_tot=np.trapz(dQ_dE, E_MeV)
            electron_total_KE_tmp=np.trapz(dQ_dE*(E_MeV/0.511-1.0), E_MeV/0.511)/Q_tot
            electron_total_KE_nulls.append(electron_total_KE_tmp)
            Qgamma2_nulls_tmp=np.trapz(dQ_dE*E_MeV/0.511*E_MeV/0.511, E_MeV)
            Qgamma2_nulls.append(Qgamma2_nulls_tmp)
            av_gamma2_hits.append(Qgamma2_nulls_tmp/Q_tot)
            peakiness_nulls, energy_thresh_tmp, max_height_nulls=compute_peakiness_measure(dQ_dE, E_MeV, color[0], normalisation_factor_E)
            Q_energy_thresh_nulls.append(energy_thresh_tmp)
            peakiness_nulls_arr.append(peakiness_nulls)
            if count<6:
                plt.plot(np.full(100, energy_thresh_tmp/rand_norm), np.linspace(0.0, 15.0, 100), color=color_arr[count])
                E_GeV=E_MeV/(1000.0*rand_norm)
                dEW_DE=dQ_dE*E_GeV/np.trapz(dQ_dE*E_GeV, E_GeV)
                plt.plot(E_GeV, dEW_DE, color=color_arr[count])
                count+=1
        else:
            e_null_list.append(spec_img*np.nan)
            E_90th_percentile_nulls.append(np.nan)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Blinded Energy (a.u.)", fontsize=16)
plt.ylabel("Normalised Charge per\n Relative Energy Spread (a.u.)", fontsize=16)
```

```python
def create_3_subplot(density_2d, delay_axis, gamma_axis):#, density_gamma):

    #print(density_2d)
    electron_density_chirp_gamma=np.trapz(density_2d, delay_axis, axis=0)#np.sum(density_2d, axis=1)/len(density_2d[0])
    electron_density_chirp_delay=np.trapz(density_2d, gamma_axis, axis=1)

    figure = plt.figure(figsize=(9,4))
    ax1 = plt.Axes(figure, [.1, .35, .10, .60])
    figure.add_axes(ax1)

    # middle
    ax2 = plt.Axes(figure, [.25, .35, .60, .60])
    figure.add_axes(ax2)

    # right
    ax3= plt.Axes(figure, [.25, .1, .60, .12])
    figure.add_axes(ax3)

    energy_axis=gamma_axis
    ax1.plot(electron_density_chirp_delay, delay_axis, color="b")
    im=ax2.pcolormesh(energy_axis, delay_axis, density_2d, shading='auto')
    ax3.plot(energy_axis, electron_density_chirp_gamma, color="r", label='pre-collision\n spectra')

    #ax1.plot(density_gamma, energy_axis, color="r", linestyle="--")

    ax3.set_xlim([min(energy_axis),1100])
    ax2.set_xlim([min(energy_axis),1100])

    ax2.set_ylim([min(delay_axis),max(delay_axis)])
    ax1.set_ylim([min(delay_axis),max(delay_axis)])

    ax3.set_xlabel("Electron energy (MeV)", fontsize=16)
    ax3.set_ylabel("Electron\n energy\n (J/GeV)", fontsize=14, rotation=270,labelpad=50)
    
    ax3.yaxis.set_label_position("right")
    ax3.yaxis.tick_right()
    #ax3.set_yticks(rotation=45)
    #ax3.yticks(rotation=45)
    
    # plt.colorbar(cb)
    # figure.colorbar(im, cax=ax2, orientation='vertical')
    # divider = make_axes_locatable(ax2)
    # cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar_ax = figure.add_axes([0.88, .30, 0.02, 0.65])
    # cax.set_ylabel('Normalised Electron Density', rotation=270)
    cb=figure.colorbar(im, cax=cbar_ax , orientation='vertical')
    cb.ax.set_ylabel('Electron energy (J/GeV/mrad)', rotation=270, labelpad=17, fontsize=14)
    cb.ax.tick_params(labelsize=14)
    ax3.tick_params(axis='both', which='major', labelsize=14)
    ax2.tick_params(axis='both', which='major', labelsize=14)
    ax1.tick_params(axis='both', which='major', labelsize=14)
    ax3.ticklabel_format(axis='y', style="scientific", scilimits=(0,0))
    #ax3.get_yaxis().get_offset_text().set_position((0.0,0))
    ax3.get_yaxis().get_offset_text().set_fontsize(14)
    ax1.set_ylabel("Divergence (mrad)", fontsize=14)
    ax1.set_xlabel("Electron\n energy (J/mrad)", fontsize=14)
    #plt.tight_layout()
    return ax1, ax2, ax3


create_3_subplot(img*E_MeV/1e3, y_mrad_disp, E_MeV)#, density_gamma)/np.mean(img)
plt.savefig("example_espec.png", bbox_inches='tight')
plt.show()
# img1=
# plt.pcolormesh(E_MeV,y_mrad_disp,img/np.mean(img)*E_MeV/1e3,shading='auto',vmin=0,vmax=15)
# plt.show()
# print("hi")
```
