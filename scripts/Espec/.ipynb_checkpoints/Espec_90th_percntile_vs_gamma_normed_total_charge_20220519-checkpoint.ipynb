```python
def select_shots(gamma_sig, s_sum, g_thresh_lo, g_thresh_hi):
 # select shots with gamma profile max counts greater than 2000
# and >50 (pc/fc?) charge
    g_max = 400000
    i_sort = np.argsort(gamma_sig)[::-1]
    gamma_sort = gamma_sig[i_sort]
    g_sel = (gamma_sort>g_thresh_lo)*(gamma_sort<g_thresh_hi)
    e_sel = np.array(s_sum)>50
    N_sel = np.sum(g_sel)
    i_sel = i_sort[g_sel]
    i_sel = i_sel[e_sel[i_sel]]
    return i_sel

i_sel_hi=select_shots(gamma_sig, s_sum, 2000, 200000)
i_sel_mid=select_shots(gamma_sig, s_sum, 1000, 2000)
i_all_null=select_shots(gamma_sig, s_sum, 0, 100)
print('n hi signal shots=%s'%len(i_sel_hi))
print('n mid signal shots=%s'%len(i_sel_mid))
print('n nulls=%s'%len(i_all_null))
```

```python
def get_nearest_nulls(i_sel, g_zero, n, nulls_per_shot):
    i_null=[]
    for ind in i_sel:
        i_diff = np.abs(n-ind)
        i_diff_sorted=np.argpartition(i_diff, nulls_per_shot)[0:nulls_per_shot]
        i_null.append(n[i_diff_sorted])
    return np.array(i_null)

#idenstify gamma null shots (where espec1 was working:)
# make array of gamma profile indices, select indices where espec has more than 50 (?) charge
# identify nulls nearest to brightest hits
g_zero = 100
n = np.arange(len(gamma_sig))[(gamma_sig<g_zero)*e_sel]
nulls_per_shot=10
i_null_hi=get_nearest_nulls(i_sel_hi, g_zero, n, nulls_per_shot)
i_null_mid=get_nearest_nulls(i_sel_mid, g_zero, n, nulls_per_shot)
```

```python
# read image using date, run and shots number (found prev)
#compute gamma profile centroid position of 99.9th percentile counts for hits
# select the part of the image that is plus/minus 200 pizels around the centroid (in x and y)
#same for nulls, but select plus/mius 120 around centroid

def get_gamma_img_tot_counts(gammaProf_db, i_sel, sel_range):
    g_img_list = []
    g_hits_total_counts=[]
    for ind in i_sel:
        row = gammaProf_db.dataframe.iloc[ind]
        run_name,shot = row.name

        file_path = get_file_path('GammaProfile',run_name,None,shot)
        img = prep_img_file(file_path)
        cx,cy = calc_COW(img,img_thresh=np.percentile(img,99.9)/2)
        y_sel = np.arange(-sel_range,sel_range)+int(cy)
        x_sel = np.arange(-sel_range,sel_range)+int(cx)
        XS,YS = np.meshgrid(x_sel,y_sel, copy=False)
        img = img[YS,XS]
        g_img_list.append(img)
        g_hits_total_counts.append(np.sum(img))
        
    return np.array(g_img_list), np.array(g_hits_total_counts)

g_img_mid_hits, g_total_counts_mid_hits=get_gamma_img_tot_counts(gammaProf_db, i_sel_mid, 200)
g_img_hi_hits, g_total_counts_hi_hits=get_gamma_img_tot_counts(gammaProf_db, i_sel_hi, 200)
```

```python
# read in gamma spec params for nulls

def get_norm_factor():
    norm_factor_file=CALIB_FOLDER+"/GammaStack/gamma_stack_nomalisation_factor.pkl"
    norm_factor_data=load_object(norm_factor_file)
    return norm_factor_data[list(norm_factor_data.keys())[0]]

def calc_Compton_energy_spec(Egamma_MeV_interp, Ec, height):#, offset):
        """
        Returns Compton spectrum given a critical energy
        """
        spec= np.exp(-Egamma_MeV_interp/Ec)*(Egamma_MeV_interp)**-(2.0/3.0)
        return spec*height#+offset
    
def get_gamma_spec_selected(i_sel, gammaProf_db):
    
    arr_shape=i_sel.shape
    e_gamma_axis=np.linspace(0.511, 920, 500)

    Ec_mean_std=[]
    height_mean_std=[]
    total_gstack_energy=[]
    gamma_norm_factor=get_norm_factor()

    for ind in i_sel.flatten():
        row = gammaProf_db.dataframe.iloc[ind]
        print(row)
        run_name,shot = row.name
        
        file_name=Path("/home/el1103292/Desktop/Analysis/GammaSpec/Bayesian_inference/"+run_name[0:9]+"GammaStack_%s.pkl"%(run_name[9:]))
        gamma_spec_data=load_object(file_name)[run_name[9:]]
        shot_keys=list(gamma_spec_data.keys())

        Ec_fitted, height_fitted=gamma_spec_data[shot][:, 1], gamma_spec_data[shot][:, 2]

        mean_Ec_blinded, mean_height_blinded=np.mean(Ec_fitted)/gamma_norm_factor, np.mean(height_fitted)/gamma_norm_factor
        std_Ec_blinded, std_height_blinded=np.std(Ec_fitted)/gamma_norm_factor, np.std(height_fitted)/gamma_norm_factor
        
        E_gamma_axis=np.outer(e_gamma_axis, np.ones(len(Ec_fitted)))
        print('E_gamma_axis.shape=%s'%[E_gamma_axis.shape])
        print('Ec_fitted.shape=%s'%[Ec_fitted.shape])
        print('height_fitted.shape=%s'%[height_fitted.shape])
        
        mean_gamma_spectrum=np.mean(calc_Compton_energy_spec(E_gamma_axis, Ec_fitted.reshape(1, -1), height_fitted.reshape(1, -1)), axis=1)
        total_energy_in_stack=np.trapz(mean_gamma_spectrum, e_gamma_axis, axis=0)
        total_gstack_energy.append(total_energy_in_stack)

        Ec_mean_std.append([mean_Ec_blinded, std_Ec_blinded])
        height_mean_std.append([mean_height_blinded, std_height_blinded])
        
    return np.array(total_gstack_energy).reshape(arr_shape[0], -1), np.array(Ec_mean_std).reshape(2, arr_shape[0], -1), np.array(height_mean_std).reshape(2, arr_shape[0], -1)

total_gstack_energy_hi_hits, Ec_mean_std_hi_hits, height_mean_std_hi_hits=get_gamma_spec_selected(i_sel_hi, gammaProf_db) 
total_gstack_energy_mid_hits, Ec_mean_std_mid_hits, height_mean_std_mid_hits=get_gamma_spec_selected(i_sel_mid, gammaProf_db) 
total_gstack_energy_nulls_hi_hits, Ec_mean_std_nulls_hi_hits, height_mean_std_nulls_hi_hits=get_gamma_spec_selected(i_null_hi, gammaProf_db) 
print("here")
total_gstack_energy_nulls_mid_hits, Ec_mean_std_nulls_mid_hits, height_mean_std_nulls_mid_hits=get_gamma_spec_selected(i_null_mid, gammaProf_db) 
       
```
