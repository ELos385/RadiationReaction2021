```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
def load_object(filename):
    with open(filename, 'rb') as fid:
        return pickle.load(fid)
    
    
prediction_test_data = load_object(r'/Volumes/Seagate Expansion Drive/Radiation_Reaction_2021/Analysis/NN_results/prediction_test_data_01.pkl')
print(list(prediction_test_data.keys()))
spectra = prediction_test_data['spectra']
pred_spectra = prediction_test_data['pred_spectra']
E_MeV = prediction_test_data['E_MeV']
```

```python
fig,axs = plt.subplots(5,4,figsize=(8,12),dpi=150)
axs = axs.flatten()
for n in range(np.shape(spectra)[0]):
    ax = axs[n]
    ax.plot(E_MeV,spectra[n,:],'k-',label='Measured')
    ax.plot(E_MeV,pred_spectra[n,:],'b-',label='Predicted')
#     ax.legend()
```
