```python
spec_info = []
for n,spec in enumerate(spec1_arr):
    spec1 = spec1_arr[n]
    y1_mrad = y_mrad_arr[n]
    E_MeV = E_MeV_arr[n]
    E,Y = np.meshgrid(E_MeV,y1_mrad)
    spec1_thresh = np.percentile(spec1,99.9,axis=0)*0.5
    spec1_AT = spec1>=(spec1_thresh.reshape(1,-1))
    y_mean = (np.sum(spec1*y1_mrad.reshape(-1,1)*spec1_AT,axis=0)/np.sum(spec1*spec1_AT,axis=0))
    # np.max(np.)
    Y_diff = Y-y_mean.reshape(1,-1)
    E_sel = spec1_thresh>=(np.std(spec1)+np.median(spec1))
    y_FWHM = np.max((Y_diff)*spec1_AT,axis=0)-np.min((Y_diff)*spec1_AT,axis=0)
    spec_max = np.percentile(spec1,99.99)
    
    spec_prj = np.sum(spec1,axis=1)
    spec_prj_max = np.max(spec_prj)
    spec_prj_FWHM = np.max(y1_mrad*(spec_prj>=spec_prj_max/2))-np.min(y1_mrad*(spec_prj>=spec_prj_max/2))
    
    spec_info.append(dict(E_axis = E_MeV[E_sel],spec_max = spec_max,y_FWHM = y_FWHM[E_sel],spec_prj_FWHM=spec_prj_FWHM))
    
    
#     plt.plot(E_MeV[E_sel],y_FWHM[E_sel])
```

```python
import scipy.optimize as opt

def Poder_div(E_MeV, c1, c2, c3, c4):#c3,
    me_GeV=0.5109989461*10**(-3)
    gamma=E_MeV/(me_GeV*1000.0)
    gamma_norm=gamma*me_GeV
    return c1+c2*gamma_norm+c3*gamma_norm**2+c4/gamma_norm#(c1*(1.0+gamma_norm)*np.log(c2*gamma_norm+1.0))**c4+c3#c4*gamma_norm**2#c1/((gamma_norm-c2)**2)+c3*(gamma_norm-c2)#c1+c2*np.log(gamma**2*me_GeV*c3)#c1*np.exp(-gamma**2*me_GeV*c2)+c3

from scipy.interpolate import interp1d

s_max = []
y_FWHM_med = []
y_FHWM_prj = []
y_FWHM_list = []
plt.figure(figsize=(4,3),dpi=300)
for n,s in enumerate(spec_info):
    s_max.append(s['spec_max'])
    if s['spec_max']>0.05:
        y_f = interp1d(s['E_axis'],s['y_FWHM'],bounds_error=False,fill_value=np.nan)(E_MeV)
        y_FWHM_list.append(y_f)
        plt.plot(s['E_axis'],s['y_FWHM'],'r-',alpha=0.1)
        y_FWHM_med.append(np.median(s['y_FWHM']))
        y_FHWM_prj.append(s['spec_prj_FWHM'])
plt.plot(E_MeV,np.nanmedian(y_FWHM_list,axis=0),'k-',label='median')
y_med = np.nanmedian(y_FWHM_list,axis=0)
i_sel = np.isfinite(y_med)
E_MeV_sel=E_MeV[i_sel]
y_med_sel=y_med[i_sel]
E_max_ind=np.argmin(abs(E_MeV_sel-1250))
print(np.argmin(abs(E_MeV_sel-1250)))
P = np.polyfit(E_MeV_sel[:E_max_ind]/1e3, y_med_sel[:E_max_ind],2)
popt, pcov=opt.curve_fit(Poder_div, E_MeV_sel[:E_max_ind], y_med_sel[:E_max_ind], p0=[1.0, 1.0, 1.0, 1.0], maxfev=10000)#, 1.0
div_fitted_exp=Poder_div(E_MeV, popt[0], popt[1], popt[2], popt[3])
plt.plot(E_MeV,np.polyval(P,E_MeV/1e3),'b--',label='Linear fit.')
plt.plot(E_MeV,div_fitted_exp,'g--',label='exp fit.')

# plt.fill_between(np.st(y_FWHM_list,axis=0),'k-')
print(P)
print("popt=%s"%popt)
plt.legend(loc=1)
plt.ylim([0,3])
# plt.xlim([400,1200])
plt.xlabel(r'$E$ [MeV]')
plt.ylabel(r'$\Delta y_{{FWHM}}$ [mrad]')
plt.tight_layout()
plt.savefig('ElectronDivergence.png',dpi=300)
```
