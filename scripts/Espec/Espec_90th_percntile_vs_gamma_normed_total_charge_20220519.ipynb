```python
# reads gamma profile analysis
analysis_path = Path(r'/data/analysis/GEMINI/2021/App20110008-1/Results/')
diag = 'GammaProfile'
database_file = analysis_path / f'{diag}_results_main.pkl'
gammaProf_db = Database(database_file,always_load=True)
gammaProf_db.dataframe.head()
```

```python
# reads processed electron spectra
s_sum = []
for n,row in gammaProf_db.dataframe.iterrows():
    run_name,shot = row.name
    spectra_path = Path('/data/analysis/GEMINI/2021/App20110008-1/Results/Espec/Spectra') / run_name.replace('/','_')
    espec_file = spectra_path / f'Shot{shot:04}_spectra.pkl'
    if os.path.isfile(espec_file):
        espec_data = load_object(espec_file)
        #print(espec_data.keys())
        E_MeV,y_mrad,spec1 = espec_data['E_MeV'],espec_data['y1_mrad'],espec_data['spec1']
        #print(len(espec_data['dWdE1_JperGeV']))
        s_sum.append(np.trapz(np.trapz(spec1,x=y_mrad,axis=0),x=E_MeV))
    else:
        s_sum.append(0.0)
```

```python
# plot total electron charge vs shot number (cumulative)
plt.figure(figsize=np.array([8,4])/2.54,dpi=150)
plt.plot(s_sum,'k-',lw=.1)
plt.xlabel('Shot')
plt.ylabel('Electron beam charge')
```

```python
# plot max gamma profile counts above background, scaled by fractional electron charge
print(s_sum/np.max(s_sum))
gamma_sig = gammaProf_db.dataframe['img_max'].values-gammaProf_db.dataframe['img_med'].values#/(s_sum/np.max(s_sum))
plt.plot(gamma_sig)
```

```python
#sort gamma profile signal by max counts, highest to lowest
# edit this to divide by total charge before sorting
i_sort = np.argsort(gamma_sig)[::-1]
plt.plot(gamma_sig[i_sort])
```

```python
# select shots with gamma profile max counts greater than 2000
# and >50 (pc/fc?) chargen
g_thresh = 1000
g_max = 400000
gamma_sort = gamma_sig[i_sort]
g_sel = (gamma_sort>g_thresh)*(gamma_sort<g_max)
e_sel = np.array(s_sum)>50
N_sel = np.sum(g_sel)
i_sel = i_sort[g_sel]
i_sel = i_sel[e_sel[i_sel]]
N_sel = len(i_sel)
N_sel
```

```python
g_zero = 100
#idenstify gamma null shots (where espec1 was working:)
# make array of gamma profile indices, select indices where espec has more than 50 (?) charge
# and the gamma signal is less than the null threshold (100 counts)
# identify nulls nearest to brightest hits
n = np.arange(len(gamma_sig))[(gamma_sig<g_zero)*e_sel]
i_null=[]
for ind in i_sel:

    i_diff = np.abs(n-ind)
    i_null.append(n[np.argmin(i_diff)])

i_null
```

```python
from PIL import Image
from scipy.ndimage import median_filter

def load_img(file_path):
    return np.array(Image.open(file_path)).astype(float)
bkg_img = 0
def prep_img_file(file_path):
    img = load_img(file_path)-bkg_img
    img = median_filter(img,9)
    
    return img
```

```python
#read date, run and shot number of brightest collisions
for ind in i_sel:
    row = gammaProf_db.dataframe.iloc[ind]
    run_name,shot = row.name
    print(row.name)
```

```python
# read image using date, run and shots number (found prev)
#compute gamma profile centroid position of 99.9th percentile counts for hits
# select the part of the image that is plus/minus 200 pizels around the centroid (in x and y)
#same for nulls, but select plus/mius 120 around centroid
g_img_list = []
for ind in i_sel:
    row = gammaProf_db.dataframe.iloc[ind]
    run_name,shot = row.name

    file_path = get_file_path(diag,run_name,None,shot)
    img = prep_img_file(file_path)
    cx,cy = calc_COW(img,img_thresh=np.percentile(img,99.9)/2)
    y_sel = np.arange(-200,200)+int(cy)
    x_sel = np.arange(-200,200)+int(cx)
    XS,YS = np.meshgrid(x_sel,y_sel)
    img = img[YS,XS]
    g_img_list.append(img)

g_null_list = []
for ind in i_null:
    row = gammaProf_db.dataframe.iloc[ind]
    run_name,shot = row.name

    file_path = get_file_path(diag,run_name,None,shot)
    img = prep_img_file(file_path)
    cx,cy = calc_COW(img,img_thresh=np.percentile(img,99.9)/2)
    y_sel = np.arange(-120,120)+int(cy)
    x_sel = np.arange(-120,120)+int(cx)
    XS,YS = np.meshgrid(x_sel,y_sel)
    img = img[YS,XS]
    g_null_list.append(img)
```

```python
# read electron divergence and spectrum (and energy axis) for hits
# computes interpolating spline/smoothing function (to remove effect of wires) across spec and div (1d) data.
# shifts divergence so that centre lies at 0

def compute_percentile_from_dist(dist, axis, target_percentile):
    
    div=10.0
    N=int(len(dist)/div)
    percentile, energy=np.zeros(N), np.zeros(N)
    dist=dist/np.trapz(dist, axis)
    for i in range(0, N):
        percentile[i]=np.trapz(dist[0:len(axis)-1-int(div)*i], axis[0:len(axis)-1-int(div)*i])
        energy[i]=axis[len(axis)-1-int(div)*i]
        if percentile[i]<target_percentile-0.005:
            break
    percentile_cut=percentile[percentile!=0.0]
    energy_cut=energy[energy!=0.0]
    energy_at_percentile=np.interp(target_percentile, percentile_cut, energy_cut)
    return energy_at_percentile
    
e_img_list = []
E_90th_percentile_hits=[]
for ind in i_sel:
    row = gammaProf_db.dataframe.iloc[ind]
    run_name,shot = row.name

    spectra_path = Path('/data/analysis/GEMINI/2021/App20110008-1/Results/Espec/Spectra') / run_name.replace('/','_')
    espec_file = spectra_path / f'Shot{shot:04}_spectra.pkl'
    if os.path.isfile(espec_file):
        espec_data = load_object(espec_file)
        E_MeV,y_mrad,spec1, dWdE1_JperGeV = espec_data['E_MeV'],espec_data['y1_mrad'],espec_data['spec1'],espec_data['dWdE1_JperGeV']
        y0 = y_mrad[np.argmax(np.sum(spec1,axis=1))]
        y_mrad_disp = np.linspace(-4.5,4.5,100)
        spec_img = RectBivariateSpline(y_mrad-y0,E_MeV,spec1)(y_mrad_disp,E_MeV)
        e_img_list.append(spec_img)
        e_at_90th_percentile=compute_percentile_from_dist(dWdE1_JperGeV, E_MeV/1000.0, 0.9)
        E_90th_percentile_hits.append(e_at_90th_percentile)
    else:
        e_img_list.append(spec_img*np.nan)
        E_90th_percentile_hits.append(np.nan)
# and for nulls
e_null_list = []
E_90th_percentile_nulls=[]
for ind in i_null:
    row = gammaProf_db.dataframe.iloc[ind]
    run_name,shot = row.name

    spectra_path = Path('/data/analysis/GEMINI/2021/App20110008-1/Results/Espec/Spectra') / run_name.replace('/','_')
    espec_file = spectra_path / f'Shot{shot:04}_spectra.pkl'
    if os.path.isfile(espec_file):
        espec_data = load_object(espec_file)
        E_MeV,y_mrad,spec1,dWdE1_JperGeV = espec_data['E_MeV'],espec_data['y1_mrad'],espec_data['spec1'],espec_data['dWdE1_JperGeV']
        y0 = y_mrad[np.argmax(np.sum(spec1,axis=1))]
        y_mrad_disp = np.linspace(-4.5,4.5,100)
        spec_img = RectBivariateSpline(y_mrad-y0,E_MeV,spec1)(y_mrad_disp,E_MeV)
        e_null_list.append(spec_img)
        e_at_90th_percentile=compute_percentile_from_dist(dWdE1_JperGeV, E_MeV/1000.0, 0.9)
        E_90th_percentile_nulls.append(e_at_90th_percentile)
    else:
        e_null_list.append(spec_img*np.nan)
        E_90th_percentile_nulls.append(np.nan)
```
